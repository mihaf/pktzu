# UPORABA CHESTER STEP TESTA ZA ANALIZO STABILNOSTI TELESA S POMOČJO PAMETNE MOBILNE NAPRAVE
## UVOD
Chester test stopanja na stopnico (CST) je test za merjenje kardiorespiratorne zmogljivosti. Vključuje vzpenjanje in spuščanje po stopnici določene višine pri določeni frekvenci stopanja. Ob tem se  spremlja odziv srčnega utripa na obremenitev. Test je bil potrjen za uporabo pri bolnikih s kronično obstruktivno pljučno boleznijo (KOPB), prav tako so različne študije ugotovile visoko zanesljivost in ponovljivost testa. CST meri več komponent, vključno s srčnim utripom (HR), subjektivnim naporom (RPE), ocenjuje največjo srčno frekvenco glede na starost (HRmax) in ocenjuje porabo kisika (VO2) pri vsaki fazi testa (Nascimento et al., 2024).

### PREGLED DOSEDANJE LITERATURE IN UPORABE PAMETNIH MOBILNIH NAPRAV
Številne študije poročajo visoko zanesljivost meritev, pridobljenih z uporabo žiroskopov na pametnih telefonih, ki so prepoznani kot natančen senzor za oceno posturalne stabilnosti (Alberts et al., 2015), pokazatelj tveganja padcev pri starejših (Najafi et al., 2002), rehabilitacije bolnikov z vrtoglavico in motnjami ravnotežja (Gawronska et al., 2020), bolnikov po možganski kapi (Hou et al., 2018), posameznikov s Parkinsonovo boleznijo (Sica et al., 20121) in drugimi zdravstvenimi stanji. Chester test stopanja na stopnico so raziskovalci uporabili pri bolnikih z abdominalnim rakom (Nascimento et al., 2024). S pomočjo senzorjev in žiroskopa so beležili inercijske premike križnice, z napravo Polar Beat pa so merili srčni utrip med izvajanje submaksimalnega šeststopenjskega testa pri bolnikih v predoperativnem obdobju. Cilj je bil ugotoviti, ali lahko predoperativni rezultat Chester Step testa z uporabo žiroskopa in ocenjevanjem VO2max, napove dolžino bolnišničnega zdravljenja v pooperativnem obdobju (Nascimento et al., 2024).
Ugotovili so povezavo med daljšo dolžino bolnišničnega zdravljenja (> 30 dni) po operaciji in nižjim RMS (angl. root mean square – povprečna amplituda sinusnega signala) signala med testom (Nascimento et al., 2024). Žiroskop je bil najnatančnejši pokazatelj dolžine pooperativnega zdravljenja pri tistih, ki so opravili le 1. stopnjo testa; VO2max  v tem primeru ni bil zadosti natančen pokazatelj. Visok VO2max  je dober pokazatelj dolžine pooperativnega zdravljenja pri tistih, ki so opravili 3. in 4. stopnjo testa. Največja koničasta moč RMS amplitude je statistično značilno (p < 0,0001) naraščala do 4. stopnje testa, medtem ko se je pri 5. stopnji testa statistično značilno zmanjšala (p < 0,0001). Do 3. stopnje testa le z vrednostjo VO2max ni mogoče predvideti podaljšano pooperativno zdravljenje v bolnišnici (Nascimento et al., 2024). 

### PROTOKOL IZVEDBE
Chester test stopanja velja za test aerobne zmogljivosti. Še nobena raziskava ni ugotavljala povezave med simetrijo koraka in spremembno posturalne stabilnosti ob povečevanju obremenitev med testom. V seminarski nalogi bo tako primerjana simetrija koraka leve in desne noge, prav tako pa bomo spremljali spremembo v stabilnosti telesa ob različnih utripih metronoma. Sklepamo lahko, da se ob hitrejšem stopanju, pospeški telesa povečajo, zmanjša pa se zmožnost stabilizacije telesnega središča. Pri hitrejših utripih metronoma pričakujemo povečano gibanje medenice v smeri naprej-nazaj in v smeri levo-desno. Večje gibanje medenice v lateralni smeri (levo-desno) bi lahko nakazovalo manjšo stabilnost telesa in slabše ravnotežje. S pomočjo pametne merilne naprave (mobilni telefon) bodo v aplikaciji PhyPhox izmerjeni pospeški telesa v x, y in z smeri. 

Pred začetkom preizkusa zabeležimo srčni utrip (HR) preiskovanca v mirovanju. Nato izračunamo 80 % največjega srčnega utripa (HRmax) po formuli 220 – starost ter podatek zabeležimo. Preiskovanca pred izvedbo testa seznanimo s protokolom meritve, ki je opisan v nadaljevanju. 

#### Navodila preiskovancu
Chester test stopanja na stopnico je zasnovan za merjenje vaše aerobne zmogljivosti. Pozvani boste, da stopate na stopnico in z nje z določeno frekvenco. Za izvedbo meritve boste imeli dva pametna telefona pritrjena na levi in desni kolčni sklep (na vsaki strani enega). Telefon bo postavljen v navpični smeri in fiksiran s trakom. Stopali boste gor in dol na stopnico, visoko 30 cm, v ritmu metronoma. Ob vsaki drugi minuti bo zabeležen vaš srčni utrip in stopnja napora, hkrati bo povečana frekvenca utripanja metronoma. Korak stopanja na stopnico si mora slediti v zaporedju gor-gor-dol-dol. Če želite, lahko zamenjate vodilno nogo, vendar morate slediti utripu metronoma. Test se bo tako nadaljeval postopno, dokler vaš srčni utrip ne doseže približno 80 % maksimuma ali dokler ne začutite, da je intenzivnost zmerno močna. Po končanem testu se ustavitve in naredite nekaj korakov po prostoru. Če se kadar koli počutite preutrujeni, brez sape ali omotični, prenehajte z izvajanjem testa.




## GRAFIČNI PRIKAZ

Prikaz meritve pospeškov pri utripih metronoma 60 utripov/min

```python
import pandas as pd
import matplotlib.pyplot as plt

# Initial plot display settings
initial_min = 15  # start
initial_max = 140 # end

# Load the data
file_path_right = 'data/Accelerometer_right.csv'
file_path_left = 'data/Accelerometer_left.csv'
data_right = pd.read_csv(file_path_right)
data_left = pd.read_csv(file_path_left)

data_filtered_right = data_right[data_right['Time (s)'] >= 0]
data_filtered_left = data_left[data_left['Time (s)'] >= 0]

# Make sure to reassign both time and acceleration data from the filtered dataset
full_time_data_right = data_filtered_right['Time (s)']
full_x_acceleration_data_right = data_filtered_right['Acceleration x (m/s^2)']
full_y_acceleration_data_right = data_filtered_right['Acceleration y (m/s^2)']
full_z_acceleration_data_right = data_filtered_right['Acceleration z (m/s^2)']

full_time_data_left = data_filtered_left['Time (s)'] + 2.3
full_x_acceleration_data_left = data_filtered_left['Acceleration x (m/s^2)']
full_y_acceleration_data_left = data_filtered_left['Acceleration y (m/s^2)']
full_z_acceleration_data_left = data_filtered_left['Acceleration z (m/s^2)']

# Create the figure and axis
fig, ax = plt.subplots(3, 1, figsize=(10, 5))
plt.subplots_adjust(bottom=0.5)

# Plotting the initial data segment
line, = ax[0].plot(full_time_data_right, full_x_acceleration_data_right, color='#3c97da', linestyle='--', label='Right')
ax[0].plot(full_time_data_left, full_x_acceleration_data_left, color='#dc6e55', linestyle='--', label='Left')
ax[0].set_xlim(initial_min, initial_max)
ax[0].set_ylim(-10, 10)
ax[0].set_ylabel('Acc X (m/s^2)')

line, = ax[1].plot(full_time_data_right, full_y_acceleration_data_right, color='#3c97da', linestyle='--', label='Right')
ax[1].plot(full_time_data_left, full_y_acceleration_data_left, color='#dc6e55', linestyle='--', label='Left')
ax[1].set_xlim(initial_min, initial_max)
ax[1].set_ylim(0, 40)
ax[1].set_ylabel('Acc Y (m/s^2)')

line, = ax[2].plot(full_time_data_right, full_z_acceleration_data_right, color='#3c97da', linestyle='--', label='Right')
ax[2].plot(full_time_data_left, full_z_acceleration_data_left, color='#dc6e55', linestyle='--', label='Left')
ax[2].set_xlim(initial_min, initial_max)
ax[2].set_ylim(-5, 12)
ax[2].set_ylabel('Acc Z (m/s^2)')


# Adding lines to indicate the detected steps
plt.xlabel('Time (s)')
plt.legend()
plt.tight_layout()
plt.show()
```

## UGOTOVITVE

Pospeški se pri stopanju na stopnico pojavljajo v smeri naprej-nazaj (X os), gor-dol (Y os) in levo-desno (Z os). 

Iz prvega grafa (60 utripov/min) je razvidno precejšnje prekrivanje v pospeških leve in desne noge v vseh oseh gibanja. Iz grafa lahko razpoznamo vzorec stopanja na stopnico z eno nogo in nato dostopa druge noge ter spust s stopnice, ki se v prvih dveh minutah dobro ohranja. Utripanje metronoma v ritmu 60 utripov/min tako ne vpliva na spremembe v dinamičnem ravnotežju. 

Drugi graf prikazuje pospeške telesa pri utripanju metronoma v ritmu 100 utripov/min. Opazimo lahko povečano frekvenco dogodkov, ki se nanizajo v vseh oseh gibanja. Pospeški na X osi (smer naprej-nazaj) se poveča, prav tako se povečajo tudi pospeški na Z osi (smer levo-desno). Na Z osi opazimo zanimivost. Vzorec pospeškov desne noge, ki je ves čas stopala na stopnico, ostaja približno enak, kar pomeni, da je preiskovanec dobro stabiliziral medenico ob dostopu. Vzorec pospeškov leve noge pa se na grafu razlikuje, kar nakazuje na slabšo stabilizacijo leve strani medenice ob spustu s stopnice navzdol. Z grafičnim prikazom tako lahko opazimo slabše ravnotežje pri stoji na levi nogi (in spustu desne noge navzdol) pri hitrejšem utripanju metronoma. Predvidevamo lahko, da bi, pri še hitrejših utripih metronoma (120, 140), grafični prikaz pokazal še večja odstopanja, s čimer bi lahko nakazoval posameznikovo dinamično ravnotežje pri funkcijskih dejavnostih kot je npr. stopanje na stopnico. Test smo zaključili pri 100 utripov/min, saj je preiskovanec že dosegel 80 % svojega maksimalnega srčnega utripa (sicer bi po protokolu sledili še dve stopnji: stopanje pri 120 in 140 utripih/min – kar bi bilo sicer zanimivo pogledati).

Tretji graf prikazuje spremembo pospeškov iz 80 utripov/min na 100 utripov/min. Opazimo lahko, da se simetričen, a obrnjen vzorec stopanja leve in desne noge pojavlja na X osi. Iz tega lahko sklepamo, da večja kot je simetrija vzorca stopanja, boljše je posameznikovo dinamično ravnotežje. Na Y osi (smer gor-dol) opazimo različno velike pospeške stopanja na desno nogo ob dostopu; najverjetneje zato, ker je preiskovanec še lovil ritem. Na Z osi je nekoliko nihanja v smeri levo-desno, vendar sta vzorca na začetku prehoda večinoma podobna. Večje nihanje v smeri levo-desno opazimo pri daljšem času stopanja na stopnico (drugi graf), kar kaže na utrujanje mišic, zaradi česar je prisotno več nihanja medenice in slabše dinamično ravnotežje, predvsem pri spustu z leve noge. 

Pametno mobilno napravo (mobilni telefon) bi lahko uporabili tudi pri drugih testih stopanja, saj gre za objektivno, diskretno in neprekinjeno pridobivanje podatkov o telesni dejavnosti z nizkimi stroški in visoko mero praktičnosti (Nascimento et al., 2024). Primerna je tudi za merjenje pospeškov in posredno ugotavljanje ravnotežja posameznika, kar smo dokazali z našim poskusom.
